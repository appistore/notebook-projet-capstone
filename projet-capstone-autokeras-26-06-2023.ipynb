# Loader le dataset

```python
df = pd.read_csv('../Datasets_sauvegardes/dataset-fraude-detection-fusionner-sans-nan-feature-ingeniering-18-06-2023.csv')
```

```python
pd.set_option('display.max_columns', None)
df.head(5)
```

```python
df_copie = df.copy()
```

```python
df_copie.shape
```

```python
# Separer X et y
y = df_copie['isFraud']
X = df_copie.drop(['isFraud'], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_sc = scaler.fit_transform(X)
```

```python
#train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_sc, y, random_state=1, test_size=0.20, stratify = y)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
#over_sampling sur les donnees d'entrainement

from imblearn.over_sampling import SMOTE
sm = SMOTE()

X_train, y_train = sm.fit_resample(X_train, y_train)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# AutoKeras

```python

# Initialiser le classificateur AutoKeras
clf = ak.StructuredDataClassifier(max_trials=10, objective='val_accuracy')

# Rechercher le meilleur modèle
history = clf.fit(X_train, y_train, epochs=10)

# Évaluer les performances du modèle sur l'ensemble de test
results = clf.evaluate(X_test, y_test)
print("Précision :", results)

# Obtenir la meilleure architecture trouvée
best_model = clf.export_model()

# Sauvegarder le modèle
# best_model.save('chemin/vers/votre/modele')
```

```python
# Sauvegarder le modèle
best_model.save('../Models_sauvegardes/modele_autokeras_29-06-2023')
```

```python
y_pred_autokeras = clf.predict(X_test)
y_pred_autokeras
```

```python
print('Precision: %.3f' % precision_score(y_test, y_pred_autokeras))
print('Recall: %.3f' % recall_score(y_test, y_pred_autokeras))
print('Accuracy: %.3f' % accuracy_score(y_test, y_pred_autokeras))
print('F1 Score: %.3f' % f1_score(y_test, y_pred_autokeras))
from sklearn import metrics
print('AUC = %.2f' % roc_auc_score(y_test, y_pred_autokeras))  #1
```

```python
# Définition d'un seuil de classification (par exemple 60%)
threshold = 0.6

# Labels prédits pour un seuil donné
label_pred_threshold = (y_pred_autokeras > threshold).astype(int)
# Calcul de la matrice de confusion pour un seuil donné
matrix_threshold = confusion_matrix(
    y_true=y_test, y_pred=label_pred_threshold
)

print(matrix_threshold)
```

```python
# Extract the true positive, true negative, false positive, and false negative values from the confusion matrix
tp, fp, fn, tn = matrix_threshold.ravel()
 
# Print the true positive, true negative, false positive, and false negative values
print("True Positive (TP): ", tp)
print("True Negative (TN): ", tn)
print("False Positive (FP): ", fp)
print("False Negative (FN): ", fn)
```

```python
def plot_learning_curves(history):
    # Extraire les métriques d'apprentissage du modèle
    train_loss = history.history['loss']
    val_loss = history.history['val_loss']
    train_accuracy = history.history['auc']
    val_accuracy = history.history['val_auc']

    # Tracer les courbes de perte
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(train_loss, label='Training Loss')
    plt.plot(val_loss, label='Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Training and Validation Loss')
    plt.legend()

    # Tracer les courbes d'exactitude
    plt.subplot(1, 2, 2)
    plt.plot(train_accuracy, label='Training AUC')
    plt.plot(val_accuracy, label='Validation AUC')
    plt.xlabel('Epoch')
    plt.ylabel('AUC')
    plt.title('Training and Validation AUC')
    plt.legend()

    # Afficher le graphique
    plt.tight_layout()
    plt.show()
```
